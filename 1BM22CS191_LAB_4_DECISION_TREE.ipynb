```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

drug_df = pd.read_csv('drug.csv')

print(drug_df.head())


# Convert categorical features to numerical
for column in drug_df.columns[:-1]:
    if drug_df[column].dtype == 'object':
        drug_df[column] = pd.factorize(drug_df[column])[0]

X_drug = drug_df.iloc[:, :-1]  # Features
y_drug = drug_df.iloc[:, -1]   # Labels

# Train-test split
X_train_drug, X_test_drug, y_train_drug, y_test_drug = train_test_split(X_drug, y_drug, test_size=0.2, random_state=42)

# Model training
drug_model = DecisionTreeClassifier()
drug_model.fit(X_train_drug, y_train_drug)

# Prediction and Evaluation
drug_pred = drug_model.predict(X_test_drug)
print('\nDrug Dataset Accuracy:', accuracy_score(y_test_drug, drug_pred))
print('Drug Confusion Matrix:\n', confusion_matrix(y_test_drug, drug_pred))

plt.figure(figsize=(16, 10))
plot_tree(drug_model, feature_names=X_drug.columns, class_names=[str(cls) for cls in y_drug.unique()], filled=True)
plt.title("Decision Tree for Drug Dataset")
plt.show()
```
